```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
cohort = pd.read_csv('final_cohort_dod.csv')
print(cohort.shape)
cohort['male'] = np.where(cohort['gender'] == 'M', True, False)
cohort.drop('gender', axis=1, inplace=True)
cohort.dropna(inplace=True)
print(cohort.shape)
```

```python
print(cohort.columns)
cohort.head()
print(cohort.survival_status.value_counts()/cohort.shape[0])
```

```python
# 1. Remove rows with null blood_glucose values
cohort = cohort.dropna(subset=['blood_glucose'])
# 2. Fill null values with 0 for specific columns
columns_to_fill_with_zero = ['glucose_variability', 'SOFA', 'apsiii', 'charlson_comorbidity_index', 'survival_status']
cohort[columns_to_fill_with_zero] = cohort[columns_to_fill_with_zero].fillna(0)
# 3. Fill null values in the wbc column with the mean value
mean_wbc = cohort['wbc'].mean()
cohort['wbc'].fillna(mean_wbc, inplace=True)
mean_bmi = cohort['bmi'].mean()
cohort['bmi'].fillna(mean_bmi, inplace=True)
```

```python
cohort.to_csv('final_cohort_dod_cleaned.csv', index=False)
```
